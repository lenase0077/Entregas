At beginning, I was planning on deleting the null data. However, I noticed that the BPM column had a significant amount of missing values (107/736). Instead of dropping those data points, I opted to impute the missing values by taking the mean of the column. After the imputation, I will then proceed to remove any remaining null data in the dataset

```python
mean_bpm = df['BPM'].mean()
df['BPM'].fillna(mean_bpm, inplace=True)
```

Now, i proceed to the elimination of null data.

```python
df = df.dropna()
```

Now, our dataset its clean and ready for visualization.
